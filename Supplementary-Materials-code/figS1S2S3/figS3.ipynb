```python
import matplotlib.pyplot as plt
import pandas as pd

fig_name = 'Anaheim_px'

# 读取三个CSV文件
df1 = pd.read_csv(f'{fig_name}/data_before.csv')
df2 = pd.read_csv(f'{fig_name}/data_during.csv')
df3 = pd.read_csv(f'{fig_name}/data_after.csv')

df4 = pd.read_csv(f'{fig_name}/theory_0.02_0.04.csv')
df5 = pd.read_csv(f'{fig_name}/theory_recover_0.02_0.05.csv')

# 创建散点图
plt.figure(figsize=(10, 8))  # 设置图形大小

# 绘制每个数据集的散点图，使用不同颜色和标签
plt.scatter(df1.iloc[:, 1], df1.iloc[:, 2], color='#a9a9a9',marker='^', s=80)
plt.scatter(df2.iloc[:, 1], df2.iloc[:, 2], color='#666666',marker='o', s=80, zorder=10)
plt.scatter(df3.iloc[:, 1], df3.iloc[:, 2], color='#a9a9a9',marker='v', s=80)

plt.plot(df4.iloc[:, 1], df4.iloc[:, 2], color='#666666')
plt.plot(df5.iloc[:, 1], df5.iloc[:, 2], color='#a9a9a9')

# 设置X轴和Y轴范围
x_min, x_max = -1, 1
y_min, y_max = -0.1, 10
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

# 计算中位数刻度（这里用范围的中值）
x_mid = (x_min + x_max) / 2  # x 轴中值
y_mid = (0 + y_max) / 2  # y 轴中值

# 设置 x 轴和 y 轴刻度为最小值、中值和最大值
plt.xticks([x_min, x_mid, x_max], fontsize=30)
plt.yticks([0, y_mid, y_max], fontsize=30)

# 添加标题和坐标轴标签
city_name = fig_name.split('_')[0]
plt.title(f'{city_name}', fontsize=32)
plt.xlabel(r"$\mathrm{x}'$", fontsize=30)
plt.ylabel(r"$\mathrm{P}( \mathrm{x}^{\prime} )$", fontsize=30)

# 保存为 SVG 文件
plt.savefig(f'px/{fig_name}.svg', format='svg', bbox_inches='tight')

# 显示图形
plt.show()
```
