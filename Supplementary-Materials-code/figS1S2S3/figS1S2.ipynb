```python
import matplotlib.pyplot as plt
import pandas as pd

fig_name = 'Anaheim_pk'  # 

#  CSV 
df4 = pd.read_csv(f'{fig_name}/theory_gamma5_beta1.05.csv')
df5 = pd.read_csv(f'{fig_name}/theory_recover_gamma6_beta1.1.csv')

# ----------  ----------
plt.figure(figsize=(10, 8))

plt.plot(df4.iloc[:, 1], df4.iloc[:, 2], color='#666666')
plt.plot(df5.iloc[:, 1], df5.iloc[:, 2], color='#a9a9a9')

# 
x_min, x_max = -1, 100
y_min, y_max = 0, 30
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks([0, (0 + x_max) / 2, x_max], fontsize=30)
plt.yticks([0, (0 + y_max) / 2, y_max], fontsize=30)

# 
city_name = fig_name.split('_')[0]
plt.title(f'{city_name}', fontsize=32)
plt.xlabel(r"$\mathrm{k}$", fontsize=30)
plt.ylabel(r"$\mathrm{N}( \mathrm{k} )$", fontsize=30)

plt.savefig(f'pk/{fig_name}-linear.svg', format='svg', bbox_inches='tight')
plt.show()

# ---------- log-log  ----------
plt.figure(figsize=(10, 8))

plt.plot(df4.iloc[:, 1], df4.iloc[:, 2], color='#666666')
plt.plot(df5.iloc[:, 1], df5.iloc[:, 2], color='#a9a9a9')

# 
plt.xscale('log')
plt.yscale('log')
plt.xlim(0.9, 150)
plt.ylim(0.9, 50)
plt.tick_params(axis='both', labelsize=30)

# 
city_name = fig_name.split('_')[0]
plt.title(f'{city_name}', fontsize=32)
plt.xlabel(r"$\mathrm{k}$", fontsize=30)
plt.ylabel(r"$\mathrm{N}( \mathrm{k} )$", fontsize=30)

plt.savefig(f'pk/{fig_name}-log.svg', format='svg', bbox_inches='tight')
plt.show()
```
