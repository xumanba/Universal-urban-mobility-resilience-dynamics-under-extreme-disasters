```python
import matplotlib.pyplot as plt
import numpy as np
import json
import sys
import pandas as pd
import tqdm
import statistics

state2 = 'TX'
city2 = 'Houston'
date2 = ['01-25', '02-01', '02-08', '02-15', '02-22', '03-01', '03-08', '03-15', '03-22', '03-29', '04-05', '04-12', '04-19']

Edge_num2 = []
for x in tqdm.tqdm(date2):
    with open(f'{state2}_{city2}_{x}_Rij.json', 'r', encoding='utf-8') as f:
        Rij_dict = json.load(f)
    with open(f'Aij_{state2}_{city2}_{x}.json', 'r', encoding='utf-8') as f:
        Aij_dict = json.load(f)
    coordinate_list = []
    for k, v in Rij_dict.items():
        for i in range(len(v)):
            coordinate_tuple = (Rij_dict[k][i], Aij_dict[k][i])
            coordinate_list.append(coordinate_tuple)
    coordinate_list = [point for point in coordinate_list if point[0] <= 50]
    coordinate_list = [point for point in coordinate_list if point[1] > 0]
    edge_num = len(coordinate_list)
    Edge_num2.append(edge_num)

print(Edge_num2)
t_axis = list(range(len(date2)))
t_axis2 = [x + 1 for x in t_axis]

plt.figure(figsize=(10, 8))
plt.bar(t_axis2, Edge_num2, color='grey', width=0.7)
plt.show()
```
