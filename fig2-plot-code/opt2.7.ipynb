```python
import matplotlib.pyplot as plt
import pandas as pd

fig_name = 'fig2.7'

# 读取三个CSV文件

df4 = pd.read_csv(f'{fig_name}/theory_gamma10_beta1.05lambda0.01.csv')
df5 = pd.read_csv(f'{fig_name}/theory_recover_gamma10_beta1.2lambda0.01.csv')

# 创建散点图
plt.figure(figsize=(10, 8))  # 设置图形大小

# 绘制每个数据集的散点图，使用不同颜色和标签

plt.plot(df4.iloc[:, 1], df4.iloc[:, 2], color='#666666')
plt.plot(df5.iloc[:, 1], df5.iloc[:, 2], color='#a9a9a9')

# 设置X轴和Y轴范围
x_min, x_max = -1, 100
y_min, y_max = -1, 120
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

# 计算中位数刻度（这里用范围的中值）
x_mid = (0 + x_max) / 2  # x 轴中值
y_mid = (0 + y_max) / 2  # y 轴中值

# 设置 x 轴和 y 轴刻度为最小值、中值和最大值
plt.xticks([0, x_mid, x_max], fontsize=30)
plt.yticks([0, y_mid, y_max], fontsize=30)

# 保存为 PDF 文件
plt.savefig(f'fig2/{fig_name}-out.pdf', format='pdf', bbox_inches='tight')

# 显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

fig_name = 'fig2.7'

# 读取三个CSV文件

df4 = pd.read_csv(f'{fig_name}/theory_gamma10_beta1.05lambda0.01.csv')
df5 = pd.read_csv(f'{fig_name}/theory_recover_gamma10_beta1.2lambda0.01.csv')

# 创建散点图
plt.figure(figsize=(10, 8))  # 设置图形大小

# 绘制每个数据集的散点图，使用不同颜色和标签

plt.plot(df4.iloc[:, 1], df4.iloc[:, 2], color='#666666')
plt.plot(df5.iloc[:, 1], df5.iloc[:, 2], color='#a9a9a9')

# 设置坐标轴为对数刻度
plt.xscale('log')
plt.yscale('log')

plt.xlim(0.9, 80)

# 增大刻度值的字体大小
plt.tick_params(axis='both', labelsize=30)  # 'both' 表示 x 和 y 轴，labelsize 设置字体大小

# 保存为 PDF 文件
plt.savefig(f'fig2/{fig_name}-in.pdf', format='pdf', bbox_inches='tight')

# 显示图形
plt.show()
```
