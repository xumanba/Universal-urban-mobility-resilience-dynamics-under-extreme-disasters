```python
import matplotlib.pyplot as plt
import pandas as pd

fig_name = 'fig2.6'

# CSV
df1 = pd.read_csv(f'{fig_name}/data_before.csv')
df2 = pd.read_csv(f'{fig_name}/data_during.csv')
df3 = pd.read_csv(f'{fig_name}/data_after.csv')

df4 = pd.read_csv(f'{fig_name}/theory_gamma2.5_beta2.csv')
df5 = pd.read_csv(f'{fig_name}/theory_recover_gamma3_beta2.csv')

# 
plt.figure(figsize=(10, 8))  # 

# 
plt.scatter(df1.iloc[:, 1], df1.iloc[:, 2], color='#a9a9a9',marker='^', s=80)
plt.scatter(df2.iloc[:, 1], df2.iloc[:, 2], color='#666666',marker='o', s=80, zorder=10)
plt.scatter(df3.iloc[:, 1], df3.iloc[:, 2], color='#a9a9a9',marker='v', s=80)

plt.plot(df4.iloc[:, 1], df4.iloc[:, 2], color='#666666')
plt.plot(df5.iloc[:, 1], df5.iloc[:, 2], color='#a9a9a9')

# XY
x_min, x_max = -1, 100
y_min, y_max = 0, 8
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

# 
x_mid = (0 + x_max) / 2  # x 
y_mid = (0 + y_max) / 2  # y 

#  x  y 
plt.xticks([0, x_mid, x_max], fontsize=30)
plt.yticks([0, y_mid, y_max], fontsize=30)

# 
city_name = 'Hollywood'
plt.title(f'{city_name}', fontsize=32)
plt.xlabel(r"$\mathrm{k}$", fontsize=30)
plt.ylabel(r"$\mathrm{N}( \mathrm{k} )$", fontsize=30)

#  PDF 
plt.savefig(f'fig2/{fig_name}-out.svg', format='svg', bbox_inches='tight')

# 
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

fig_name = 'fig2.6'

# CSV
df1 = pd.read_csv(f'{fig_name}/data_before.csv')
df2 = pd.read_csv(f'{fig_name}/data_during.csv')
df3 = pd.read_csv(f'{fig_name}/data_after.csv')

df4 = pd.read_csv(f'{fig_name}/theory_gamma2.5_beta2.csv')
df5 = pd.read_csv(f'{fig_name}/theory_recover_gamma3_beta2.csv')

# 
plt.figure(figsize=(10, 8))  # 

# 
plt.scatter(df1.iloc[:, 1], df1.iloc[:, 2], color='#a9a9a9',marker='^', s=80)
plt.scatter(df2.iloc[:, 1], df2.iloc[:, 2], color='#666666',marker='o', s=80, zorder=10)
plt.scatter(df3.iloc[:, 1], df3.iloc[:, 2], color='#a9a9a9',marker='v', s=80)

plt.plot(df4.iloc[:, 1], df4.iloc[:, 2], color='#666666')
plt.plot(df5.iloc[:, 1], df5.iloc[:, 2], color='#a9a9a9')

# 
plt.xscale('log')
plt.yscale('log')

plt.xlim(0, 100)
plt.ylim(0, 10)

# 
plt.tick_params(axis='both', labelsize=30)  # 'both'  x  y labelsize 

# 
city_name = 'Hollywood'
plt.title(f'{city_name}', fontsize=32)
plt.xlabel(r"$\mathrm{k}$", fontsize=30)
plt.ylabel(r"$\mathrm{N}( \mathrm{k} )$", fontsize=30)

#  PDF 
plt.savefig(f'fig2/{fig_name}-in.svg', format='svg', bbox_inches='tight')

# 
plt.show()
```
